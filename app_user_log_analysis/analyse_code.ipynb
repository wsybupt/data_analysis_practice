# 探索性分析

```python
# 计算z值
n_new = new_df.shape[0]
n_old = old_df.shape[0]

import math

def calculate_z(column, bigThenZero = True):
    if bigThenZero:
        count_new =  new_df[new_df[column]>0].loc[:,column].count()
        count_old =  old_df[old_df[column]>0].loc[:,column].count()
    else:
        count_new =  new_df[new_df[column]==0].loc[:,column].count()
        count_old =  old_df[old_df[column]==0].loc[:,column].count()
    p_new = count_new/n_new
    p_old = count_old/n_old
    q_new = 1 - p_new
    q_old = 1 - p_old
    z = (p_new - p_old)/math.sqrt(p_new*q_new/n_new + p_old*q_old/n_old)
    dict1 = {'count_new':count_new, 'p_new':p_new, 'count_old':count_old, 'p_old':p_old, 'z':z}
    return dict1
```
